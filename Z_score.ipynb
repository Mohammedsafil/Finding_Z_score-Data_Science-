```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("movie_revenues.csv")
```

```python
df
```

```python
df["revenue_in_m"]=df["revenue"].apply(lambda x: x/1000000)
```

```python
df.head(2)
```

```python
mean = df.revenue_in_m.mean()
mean
```

```python
std = df.revenue_in_m.std()
std
```

```python
sns.histplot(df.revenue_in_m, kde = True )
```

```python
def get_zscore(val,mean,std):
    return (val-mean)/std
```

```python
df["z_score"] = df.revenue_in_m.apply(lambda x: get_zscore(x,mean,std))
```

```python
df.head(3)
```

```python
df[df.z_score>3]
```

```python
median_revenue = abs(df.revenue_in_m - df.revenue_in_m.median())
```

```python
mad = median_revenue.median()
```

```python
mad
```

```python
def modified_zscore(val,mad):
    return 0.6745*(val - df.revenue_in_m.median())/mad
```

```python
df["modified_zscore"] = df.revenue_in_m.apply(lambda x : modified_zscore(x,mad))
df.head(3)
```

```python
df[(df.modified_zscore> 3.5) | (df.modified_zscore< -3.5)]
```
